1. Realiza una investigación científica sobre el filtro Bayer aplicado en el procesamiento digital de imágenes, el porque se aplica y como proviene; de la misma manera realizando código lista las versiones de filtro Bayer en cv2 y aplicalos en tus imágenes seleccionadas.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

bayer_image = cv2.imread('2.png', cv2.IMREAD_GRAYSCALE)
bgr_image = cv2.cvtColor(bayer_image, cv2.COLOR_BAYER_BG2BGR)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(bayer_image, cmap='gray')
plt.title("Imagen Bayer")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB))
plt.title("Imagen Convertida a BGR")
plt.axis('off')

plt.show()
```

2. Aplica las transformaciones de YUV2RGB, RGB2YUV y viceversa a la imagen seleccionada

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('2.png', cv2.COLOR_BGR2RGB)
yuv_image = cv2.cvtColor(image, cv2.COLOR_RGB2YUV)

rgb_vicerversa = cv2.cvtColor(yuv_image, cv2.COLOR_YUV2RGB)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(image)
axes[0].set_title('Imagen Original RGB')
axes[0].axis('off')

axes[1].imshow(yuv_image[..., 0], cmap='gray') 
axes[1].set_title('Imagen YUV')
axes[1].axis('off')

axes[2].imshow(rgb_vicerversa)
axes[2].set_title('RGB Viceversa')
axes[2].axis('off')

plt.show()
```
