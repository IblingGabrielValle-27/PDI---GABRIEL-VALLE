PROFUNDIDAD DE COLOR

Crea una imagen en RGB que contenga 4 cuadrantes en rojo y verde, azul y negro

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np

plt.rcParams['figure.figsize']=[5,5]
tamanio_imagen = 200

R = np.zeros((tamanio_imagen, tamanio_imagen))
G = np.zeros((tamanio_imagen, tamanio_imagen))
B = np.zeros((tamanio_imagen, tamanio_imagen))

R[:tamanio_imagen//2, :tamanio_imagen//2] = 255 
G[:tamanio_imagen//2, tamanio_imagen//2:] = 255  
B[tamanio_imagen//2:, :tamanio_imagen//2] = 255  

imagen = np.dstack((R, G, B))

plt.figure()
plt.imshow(imagen)
plt.axis('off')
plt.title('Imagen con Cuadrantes Rojo, Verde, Azul y Negro')
plt.show()
```
