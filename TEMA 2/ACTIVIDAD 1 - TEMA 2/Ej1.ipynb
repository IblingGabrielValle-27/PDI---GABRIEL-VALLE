Las imagenes se encuentran registradas con T1, T2 y T3
2. Carga tus imagenes elegidas en formato BGR utilizando cv2.imread
3. Muestra cada imagen en formato BGR usando Matplotlib, pero convertida a RGB para que se muestre correctamente
4. Utiliza cv2.cvtColor para convertir la imagen de BGR a escala de grises

6. Guarda la imagen en BGR en el Ordenador

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize']=[10,5]

# Código de la imagen 1
img_bgr1= cv2.imread('T1.jpeg')
img_rgb1 = cv2.imread('T1.jpeg', cv2.IMREAD_COLOR)
    
plt.figure()
plt.imshow(cv2.cvtColor(img_rgb1, cv2.COLOR_BGR2RGB))
plt.title(f'Imagen RGB')
plt.axis('off')
plt.show()
    
img_gray = cv2.cvtColor(img_rgb1, cv2.COLOR_BGR2GRAY)
plt.figure()
plt.imshow(img_gray, cmap='gray')
plt.title(f'Escala de Grises - {img_bgr1.shape[0]}')
plt.axis('off')
plt.show()

print(f'Alto: {img_rgb1.shape[0]}')
print(f'Ancho: {img_rgb1.shape[1]}')
print(f'Canales: {img_rgb1.shape[2]}')
fig,axs=plt.subplots(1,4,figsize=(20,5))

axs[0].imshow(img_rgb1)
axs[0].set_title('imagen original')

fig_img=axs[1].imshow(img_rgb1[...,0],cmap='gray')
fig.colorbar(fig_img, ax=axs[1], shrink=0.5)
axs[1].set_title('canal rojo')

fig_img = axs[2].imshow(img_rgb1[..., 1], cmap='gray')
fig.colorbar(fig_img, ax=axs[2], shrink=0.5)
axs[2].set_title('Canal Verde')

fig_img = axs[3].imshow(img_rgb1[..., 2], cmap='gray')
fig.colorbar(fig_img, ax=axs[3], shrink=0.5)
axs[3].set_title('Canal Azul')
    
cv2.imwrite(f"{img_bgr1.shape[0]}_gris.jpg", img_gray)
#cv2.imwrite('T1.jpeg',img_bgr1)
#plt.imsave('T1.jpeg', cv2.cvtColor(img_bgr1,cv2.COLOR_BGR2RGB))

b, g, r = cv2.split(img_rgb1)
plt.figure()
    
plt.subplot(141)
plt.title('Original')
plt.imshow(cv2.cvtColor(img_rgb1, cv2.COLOR_BGR2RGB))
plt.axis('off')
    
plt.subplot(142)
plt.title('Canal Rojo')
plt.imshow(r, cmap='Reds')
plt.axis('off')
    
plt.subplot(143)
plt.title('Canal Verde')
plt.imshow(g, cmap='Greens')
plt.axis('off')
    
plt.subplot(144)
plt.title('Canal Azul')
plt.imshow(b, cmap='Blues')
plt.axis('off')


# Código de la imagen 2
img_bgr2= cv2.imread('T2.jpeg')
img_rgb2 = cv2.imread('T2.jpeg', cv2.IMREAD_COLOR)
    
plt.figure()
plt.imshow(cv2.cvtColor(img_rgb2, cv2.COLOR_BGR2RGB))
plt.title(f'Imagen RGB')
plt.axis('off')
plt.show()
    
img_gray = cv2.cvtColor(img_rgb2, cv2.COLOR_BGR2GRAY)
plt.figure()
plt.imshow(img_gray, cmap='gray')
plt.title(f'Escala de Grises - {img_bgr2.shape[0]}')
plt.axis('off')
plt.show()

print(f'Alto: {img_rgb2.shape[0]}')
print(f'Ancho: {img_rgb2.shape[1]}')
print(f'Canales: {img_rgb2.shape[2]}')
fig,axs=plt.subplots(1,4,figsize=(20,5))

axs[0].imshow(img_rgb2)
axs[0].set_title('imagen original')

fig_img=axs[1].imshow(img_rgb2[...,0],cmap='gray')
fig.colorbar(fig_img, ax=axs[1], shrink=0.5)
axs[1].set_title('canal rojo')

fig_img = axs[2].imshow(img_rgb2[..., 1], cmap='gray')
fig.colorbar(fig_img, ax=axs[2], shrink=0.5)
axs[2].set_title('Canal Verde')

fig_img = axs[3].imshow(img_rgb2[..., 2], cmap='gray')
fig.colorbar(fig_img, ax=axs[3], shrink=0.5)
axs[3].set_title('Canal Azul')
    
cv2.imwrite(f"{img_bgr2.shape[0]}_gris.jpg", img_gray)
#cv2.imwrite('T2.jpeg',img_bgr2)
#plt.imsave('T2.jpeg', cv2.cvtColor(img_bgr2,cv2.COLOR_BGR2RGB))

b, g, r = cv2.split(img_rgb2)
plt.figure()
    
plt.subplot(141)
plt.title('Original')
plt.imshow(cv2.cvtColor(img_rgb2, cv2.COLOR_BGR2RGB))
plt.axis('off')
    
plt.subplot(142)
plt.title('Canal Rojo')
plt.imshow(r, cmap='Reds')
plt.axis('off')
    
plt.subplot(143)
plt.title('Canal Verde')
plt.imshow(g, cmap='Greens')
plt.axis('off')
    
plt.subplot(144)
plt.title('Canal Azul')
plt.imshow(b, cmap='Blues')
plt.axis('off')


# Código de la imagen 3

img_bgr3= cv2.imread('T3.jpeg')
img_rgb3 = cv2.imread('T3.jpeg', cv2.IMREAD_COLOR)
    
plt.figure()
plt.imshow(cv2.cvtColor(img_rgb3, cv2.COLOR_BGR2RGB))
plt.title(f'Imagen RGB')
plt.axis('off')
plt.show()
    
img_gray = cv2.cvtColor(img_rgb3, cv2.COLOR_BGR2GRAY)
plt.figure()
plt.imshow(img_gray, cmap='gray')
plt.title(f'Escala de Grises - {img_bgr3.shape[0]}')
plt.axis('off')
plt.show()

print(f'Alto: {img_rgb3.shape[0]}')
print(f'Ancho: {img_rgb3.shape[1]}')
print(f'Canales: {img_rgb3.shape[2]}')
fig,axs=plt.subplots(1,4,figsize=(20,5))

axs[0].imshow(img_rgb3)
axs[0].set_title('imagen original')

fig_img=axs[1].imshow(img_rgb3[...,0],cmap='gray')
fig.colorbar(fig_img, ax=axs[1], shrink=0.5)
axs[1].set_title('canal rojo')

fig_img = axs[2].imshow(img_rgb3[..., 1], cmap='gray')
fig.colorbar(fig_img, ax=axs[2], shrink=0.5)
axs[2].set_title('Canal Verde')

fig_img = axs[3].imshow(img_rgb3[..., 2], cmap='gray')
fig.colorbar(fig_img, ax=axs[3], shrink=0.5)
axs[3].set_title('Canal Azul')
    
cv2.imwrite(f"{img_bgr3.shape[0]}_gris.jpg", img_gray)
#cv2.imwrite('T3.jpeg',img_bgr3)
#plt.imsave('T3.jpeg', cv2.cvtColor(img_bgr3,cv2.COLOR_BGR2RGB))

b, g, r = cv2.split(img_rgb3)
plt.figure()
    
plt.subplot(141)
plt.title('Original')
plt.imshow(cv2.cvtColor(img_rgb3, cv2.COLOR_BGR2RGB))
plt.axis('off')
    
plt.subplot(142)
plt.title('Canal Rojo')
plt.imshow(r, cmap='Reds')
plt.axis('off')
    
plt.subplot(143)
plt.title('Canal Verde')
plt.imshow(g, cmap='Greens')
plt.axis('off')
    
plt.subplot(144)
plt.title('Canal Azul')
plt.imshow(b, cmap='Blues')
plt.axis('off')
    
plt.show()
```
